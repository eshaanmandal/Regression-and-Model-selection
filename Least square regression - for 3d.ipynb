```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize'] = [10,10]
# importing the necessary docs
```

```python
x = np.linspace(2,10,30)
y = np.linspace(2,4,30)

z = 2*x + 3*y + np.random.choice(np.arange(1,100))*np.random.randn(30)
```

```python
# The matrix required for determining the best fit coeffs is a symmetric one
```

```python
ak2 = np.sum(np.power(x,2))
bk2 = np.sum(np.power(y,2))
ak = np.sum(x)
bk = np.sum(y)
akbk = np.sum(x*y)
n = len(x)
A = np.array([[ak2, akbk, ak],[akbk, bk2, bk],[ak, bk, n]])
b = np.array([[np.sum(x*z)],[np.sum(y*z)],[np.sum(z)]])
s = np.linalg.solve(A,b)
```

```python
z_f = s[0]*x + s[1]*y + s[2]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x,y,z,'o')
ax.plot(x,y,z_f)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data points and best fit line')
plt.show()
```
